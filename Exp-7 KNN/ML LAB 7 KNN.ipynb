```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names=['sepallength','sepalwidth','patellength','patelwidth','class']
dataset=pd.read_csv (url,names=names)
```

```python
dataset
```

```python
trainingclass=dataset.values[:,-1]
trainingclass
```

```python
unique_list=list(trainingclass)
```

```python
for i in range (len(trainingclass)):
  for j in range (len(unique_list)):
    if(trainingclass[i]==unique_list[j]):
      trainingclass[i]=j
trainingclass
```

```python
import math

def euclidean_distance(rowsi,rowsj):
  distance=0.0
  for i in range(len(rowsi)):
    distance+=(rowsi[i]-rowsj[i])**2
  return (math.sqrt(distance))
```

```python
training=dataset.values[:,0:4]
```

```python
testing=dataset.values[149,0:4]
```

```python
distance=[]
for i in range (len(training)):
  distance.append([euclidean_distance(training[i],testing),trainingclass[i]])
```

```python
distance.sort()
```

```python
k=5
voting=[0 for i in range (len(unique_list))]
for i in range (1,k+1):
  voting[distance[i][1]]=voting[distance[i][1]]+1
```
