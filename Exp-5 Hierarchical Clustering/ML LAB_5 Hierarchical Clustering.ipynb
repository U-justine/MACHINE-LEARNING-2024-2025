```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#used for classification purpose
dataset=pd.read_csv("/content/CLUSTERING.csv")
dataset
```

```python
dataset.describe()
```

```python
#Hierarchical clustering
X=dataset.iloc[:,[1,2]].values
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
import scipy.cluster.hierarchy as sch
sch.dendrogram=sch.dendrogram(sch.linkage(X,method="ward"))
```

```python
hc=AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage='ward')
```

```python
predict_hc=hc.fit_predict(X)
```

```python
predict_hc
```

```python
#visualization
plt.scatter(X[predict_hc==0,0],X[predict_hc==0,1],c='red',label="cluster 1")
plt.scatter(X[predict_hc==1,0],X[predict_hc==1,1],c='green',label="cluster 2")
plt.scatter(X[predict_hc==2,0],X[predict_hc==2,1],c='yellow',label="cluster 3")
plt.scatter(X[predict_hc==3,0],X[predict_hc==3,1],c='blue',label="cluster 4")
plt.scatter(X[predict_hc==4,0],X[predict_hc==4,1],c='purple',label="cluster 5")
plt.title("Hierarchical Clustering")
plt.legend()
```
